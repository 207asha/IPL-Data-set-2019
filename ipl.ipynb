```python
data.groupby('season')['winner'].value_counts()
#Using group-by function we can split the data into groups based on some criteria.
#Here, we can see that the no.of matches won by the team based on the season.From,
#above syntax we can see that the maximum matches won by the Team amongst all the seasons is Rajasthan Royals.
#(Data is from 2008 to 2019, As per new reports, MI is leading match winners against the other teams in all the seasons)
```

```python
#From Match date we want to extract the dates,month,years so we use pandas date-time to do this
data['date']=pd.to_datetime(data['date'])
data['date']
#data['month']
```

```python
#Here, to extract month from match_date we use the below syntax to extract the month
data['month']=data['date'].dt.month
data['month']
```

```python
#Here, to extract year from match_date we use the below syntax to extract the month
data['year']=data['date'].dt.year
data['year']
```

```python
#Create dataframe with missing percentage
missing_val = pd.DataFrame(data.isnull().sum())
missing_val = missing_val.reset_index()

#Rename variable
missing_val = missing_val.rename(columns = {'index': 'Features', 0: 'Missing_percentage'})

#Calculate percentage
missing_val['Missing_percentage'] = (missing_val['Missing_percentage']/len(data))*100

#sort
missing_val.sort_values('Missing_percentage' )
```
